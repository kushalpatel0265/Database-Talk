```python
import os
from pandasai import Agent
from pandasai.connectors import SqliteConnector

# Set up the environment variable for the API key
os.environ["PANDASAI_API_KEY"] = "$2a$10$76r5odl8wtbCYCfv3fVJoO6jirDHHPkrPsSMXgXamVPI0JhHZXuUu"

# Create connectors for each table
albums_connector = SqliteConnector(
    config={"database": "Database.db", "table": "albums"}
)
artists_connector = SqliteConnector(
    config={"database": "Database.db", "table": "artists"}
)
customers_connector = SqliteConnector(
    config={"database": "Database.db", "table": "customers"}
)
employees_connector = SqliteConnector(
    config={"database": "Database.db", "table": "employees"}
)
genres_connector = SqliteConnector(
    config={"database": "Database.db", "table": "genres"}
)
invoice_items_connector = SqliteConnector(
    config={"database": "Database.db", "table": "invoice_items"}
)
invoices_connector = SqliteConnector(
    config={"database": "Database.db", "table": "invoices"}
)
media_types_connector = SqliteConnector(
    config={"database": "Database.db", "table": "media_types"}
)
playlist_track_connector = SqliteConnector(
    config={"database": "Database.db", "table": "playlist_track"}
)
playlists_connector = SqliteConnector(
    config={"database": "Database.db", "table": "playlists"}
)
tracks_connector = SqliteConnector(
    config={"database": "Database.db", "table": "tracks"}
)

# Initialize the agent with all connectors
agent = Agent([
    albums_connector, artists_connector, customers_connector,
    employees_connector, genres_connector, invoice_items_connector,
    invoices_connector, media_types_connector, playlist_track_connector,
    playlists_connector, tracks_connector
])
```
